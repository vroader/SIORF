```python
import pandas as pd
import sqlalchemy
import csv
```

```python
engine = sqlalchemy.create_engine('mysql+pymysql://siorf:PMseis06@mysql.siorf.kinghost.net:3306/siorf')
```

```python
tbl_cadastroDemandas = open('tbl_cadastroDemandas.csv', encoding = 'utf-8')
```

```python
tbl_cadastroDemandas_Readeble = csv.reader(tbl_cadastroDemandas, delimiter=';')
```

```python
cadastroDemandas = list(tbl_cadastroDemandas_Readeble)
```

```python
del cadastroDemandas[0]
    
```

```python
cadastroDemandas
```

```python
cadastroDemandaLimpa = []
for lista in cadastroDemandas:
    del lista[0]
    if lista[6] == 'sim':
        lista[6]= '1'
    else:
        lista[6] = '0'
    cadastroDemandaLimpa.append(lista)
cadastroDemandaLimpa
```

```python
len(cadastroDemandaLimpa)
```

```python
cadastroDemandaAtualizado=[]
for registro in cadastroDemandaLimpa:
    if registro[1]== 'CME':
        registro[1] = 'DOP'
    elif registro [1] == 'DPPHC':
        registro[1]= 'DEC'
    elif registro[1]== 'DIPRO':
        registro[1]='DINFRA'
    elif registro[1]== 'DPPP':
        registro[1]= 'DPP'
    elif registro[1] == 'DPMT':
        registro[1] = 'DPTS'
    cadastroDemandaAtualizado.append(registro)
cadastroDemandaAtualizado
```

```python
len(cadastroDemandaAtualizado)
```

```python
for registro in  cadastroDemandaAtualizado:
    engine.execute('SET FOREIGN_KEY_CHECKS=0;')
    engine.execute('INSERT INTO tbl_CadastroDemandas (cad_CodigoDemanda, cad_CoordenadorSetorial, cad_Descricao, cad_FCDFDespesaDetalhada, cad_GDFDespesaDetalhada, cad_ProdutoUnidade, cad_Periodicidade, cad_EspecieDisponibilidade) values ("'+registro[0]+'","'+registro[1]+'","'+registro[2]+'","'+registro[3]+'","'+registro[4]+'","'+registro[5]+'","'+registro[6]+'","'+registro[7]+'");')
```
