```python
engine = sqlalchemy.create_engine('mysql+pymysql://siorf:PMseis06@mysql.siorf.kinghost.net:3306/siorf')
```

```python
dfPIO = pd.read_csv('tbl_GDFPlanoInterno.csv', delimiter = ';')
dfPIO
```

```python
dfPIO['Valor'] = [x.replace('.','') for x in dfPIO['Valor']]
dfPIO['Valor'] = [x.replace(',', '.') for x in dfPIO['Valor']]
dfPIO
```

```python
propostaGDF = pd.read_sql_table('tbl_GDFProposta', engine)
propostaSetorial = pd.read_sql_table('tbl_PropostaSetorial',engine)
cadastro = pd.read_sql_table('tbl_CadastroDemandas',engine)
```

```python
cadastro.head(1)
```

```python
dfPIO1 = dfPIO.merge(cadastro, left_on='Proposta', right_on='cad_CodigoDemanda', how = 'left')
dfPIO1
```

```python
propostaSetorial.head(1)
```

```python
dfPIO2 = dfPIO1.merge(propostaSetorial, left_on='Id_CadastroDemandas', right_on='pps_CadastroDemanda', how = 'left')
```

```python
dfPIO2.head(32)
```

```python
propostaGDF.head(1)
```

```python
dfPIO3 = dfPIO2.merge(propostaGDF, left_on = 'Id_PropostaSetorial', right_on='prg_PropostaSetorial', how = 'left')
dfPIO3.head(1)
```

```python
dfPIOFINAL = dfPIO3[['Id_GDFProposta', 'Quantidade', 'Valor']]
dfPIOFINAL
```

```python
#Agora transformar o DataFrame em uma lista para ser inserido no MYSQL
list_PIO = dfPIOFINAL.values.tolist()
PIO = []
#Transformando dados 'int' em string'
for lista in list_PIO:
    linha = []
    for registro in lista:
        registro = str(registro)
        linha.append(registro)
    PIO.append(linha)
print(PIO)
```

```python
for registro in PIO:
    engine.execute('SET FOREIGN_KEY_CHECKS=0;')
    engine.execute('INSERT INTO tbl_GDFPlanoInterno ( pig_Proposta, pig_Quantidade, pig_Valor) values ("'+registro[0]+'","'+registro[1]+'","'+registro[2]+'");')
```
