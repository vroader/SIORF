```python
import pandas as pd
import sqlalchemy
import csv
```

```python
engine = sqlalchemy.create_engine('mysql+pymysql://siorf:PMseis06@mysql.siorf.kinghost.net:3306/siorf')
```

```python
PropostaSetorial = pd.read_sql_table('tbl_PropostaSetorial', engine)
IniciativaEstrategica = pd.read_sql_table('tbl_IniciativaEstrategica',engine)
```

```python
PropostaSetorial1 = PropostaSetorial.merge(IniciativaEstrategica, left_on='pps_alinhamentoEstrategico', right_on='ini_Codigo', how = 'left')
PropostaSetorial1
```

```python
PropostaSetorial2 = PropostaSetorial1[['Id_PropostaSetorial', 'Id_IniciativaEstrategica']]
PropostaSetorial2
```

```python
#Agora transformar o DataFrame em uma lista para ser inserido no MYSQL
list_PropostaSetorial = PropostaSetorial2.values.tolist()
PropostaSetorial3 = []
#Transformando dados 'int' em string'
for lista in list_PropostaSetorial:
    linha = []
    for registro in lista:
        registro = str(registro)
        linha.append(registro)
        PropostaSetorial3.append(linha)
print(PropostaSetorial3)
```

```python
for registro in PropostaSetorial3:
    engine.execute('SET FOREIGN_KEY_CHECKS=0;')
    engine.execute('UPDATE tbl_PropostaSetorial SET pps_IniciativaEstrategica = "'+registro[1]+'" WHERE Id_PropostaSetorial = "'+registro[0]+'"')
```
