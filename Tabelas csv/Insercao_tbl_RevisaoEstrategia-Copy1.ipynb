```python
import pandas as pd
import sqlalchemy
import csv
```

```python
engine = sqlalchemy.create_engine('mysql+pymysql://siorf:PMseis06@mysql.siorf.kinghost.net:3306/siorf')
```

```python
Iniciativa = open('RevisãoIniciativa.csv', encoding = 'utf-8')
```

```python
Iniciativa_Readeble = csv.reader(Iniciativa, delimiter = ';')
```

```python
listIniciativa = list(Iniciativa_Readeble)
```

```python
listIniciativa
```

```python
del listIniciativa[0]
listIniciativa
```

```python
Iniciativa1 = []
for lista in listIniciativa:
    linha = []
    codIniciativa = lista[1].split(' ')[0].strip('.')
    codEstrategia = codIniciativa[0:-2]
    descIniciativa = lista[1][(len(codEstrategia)+2):].strip(' ')
    linha.append(codIniciativa)
    linha.append(codEstrategia)
    linha.append(descIniciativa)
    Iniciativa1.append(linha)
print(Iniciativa1)
```

```python
Estrategia = pd.read_sql_table('tbl_Estrategia',engine)
Estrategia
```

```python
EstrategiaAtiva = Estrategia.loc[(Estrategia['est_Status']==1)]
```

```python
EstrategiaAtiva.head(1)
```

```python
from pandas import DataFrame
dfIniciativa = DataFrame(Iniciativa1, columns=['codIniciativa', 'codEstrategia', 'codDescricao'])
dfIniciativa
```

```python
dfIniciativaFinal = dfIniciativa.merge(EstrategiaAtiva, left_on='codEstrategia', right_on='est_Codigo', how = 'left')
dfIniciativaFinal
```

```python
dfInicitivaFinal2 = dfIniciativaFinal[['codIniciativa', 'Id_Estrategia', 'codDescricao']]
```

```python
dfInicitivaFinal2
```

```python
list_Iniciativa = dfInicitivaFinal2.values.tolist()
list_Iniciativa
```

```python
IniciativaEstrategica = []
for lista in list_Iniciativa:
    linha = []
    linha.append(lista[0].strip('.'))
    linha.append(str(lista[1]))
    linha.append(lista[2])
    IniciativaEstrategica.append(linha)
print(IniciativaEstrategica)
```

```python
for registro in  IniciativaEstrategica:
    engine.execute('INSERT INTO tbl_IniciativaEstrategica (ini_Codigo, ini_Estrategia, ini_Descricao) values ("'+registro[0]+'","'+registro[1]+'","'+registro[2]+'");')
```
