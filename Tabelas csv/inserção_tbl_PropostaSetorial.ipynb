```python
#Primeiro abro a tbl_PropostaSetorial.csv etraida do acces e tranformo em df, importante lembrar que é necessário identificar o delimitador
# Salvo o resultado em uma variável
dfProposta = pd.read_csv('tbl_PropostasSetoriais2021.csv', delimiter = ';')
dfProposta
```

```python
#Preciso remover o id_Proposta da tabaela Proposta Setorial e inserir o Código do Cadastro de Demandas do Banco em mysql
#e remover a priorização (melhor criar uma uma tabela temporária no banco pois a priorização do Plano diretor também é local)

dfPropostaLimpa = dfProposta.drop(columns=['Id_PropostaSetorial2021', 'pps_Gravidade','pps_Tendencia', 'pps_Urgencia'])
dfPropostaLimpa
```

```python
# Foi verificado que existem dois registros de diárias na Tabela Proposta Setorial e na tabela Cadstro.
# A duplicidade da tabela cadastro não pode permanecer.
# A tabela Cadastro já está inserida no Banco então o registro foi invalidade para reutilização futura
```

```python
#Abro o Cadastro com o id a ser recuperado e armazeno em uma variável
dfCadastro = pd.read_sql_table('tbl_CadastroDemandas',engine)
dfCadastro
```

```python
#Selecionar apenas as colunas que serão utilizadas para fazera correspondência entre a tabela Cadastro e a tabela proposta
colunaIdCadastro = dfCadastro[['Id_CadastroDemandas', 'cad_CodigoDemanda']]
colunaIdCadastro
```

```python
# Foi atribuído o Id_do Cadastro da demanda na Proposta Setorial
propostaMerge = pd.merge(colunaIdCadastro, dfPropostaLimpa, left_on = "cad_CodigoDemanda", right_on = "pps_CodigoDemanda", how = "right" )
```

```python
#Verificando se a Proposta esta completa, sobretudo no que tange a demanda de código 61
propostaMerge.loc[(propostaMerge['cad_CodigoDemanda']) == 61]
```

```python
#Agora é necessário remover as Colunas com Código da Demanda.
```

```python
propostaMerge = propostaMerge.drop(columns=['cad_CodigoDemanda', 'pps_CodigoDemanda'])
propostaMerge
```

```python
#Agora transformar o DataFrame em uma lista para ser inserido no MYSQL
list_PropostaSetorial = propostaMerge.values.tolist()
Proposta = []
#Transformando dados 'int' em string'
for lista in list_PropostaSetorial:
    linha = []
    for registro in lista:
        registro = str(registro)
        linha.append(registro)
    Proposta.append(linha)
print(Proposta)
```

```python
for registro in Proposta:
    engine.execute('SET FOREIGN_KEY_CHECKS=0;')
    engine.execute('INSERT INTO tbl_PropostaSetorial (pps_CadastroDemanda, pps_Justificativa, pps_AlinhamentoEstrategico, pps_Quantidade, pps_Valor, pps_Observacao, pps_UnidadeDemandante, pps_Exercicio) values ("'+registro[0]+'","'+registro[1]+'","'+registro[2]+'","'+registro[3]+'","'+registro[4]+'","'+registro[5]+'","'+registro[6]+'","'+registro[7]+'");')
```
