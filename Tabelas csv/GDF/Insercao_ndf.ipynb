```python
tbl_GDFNaturezaDespesaDetalhada = open('tbl_GDFNaturezaDespesaDetalhada.csv')
```

```python
NaturezaDespesaDetalhada_Readeble = csv.reader(tbl_GDFNaturezaDespesaDetalhada)
```

```python
NaturezaDespesaDetalhada = list(NaturezaDespesaDetalhada_Readeble)
```

```python
NaturezaDespesaDetalhada
```

```python
def is_number(num):
    try:
        float(num)
        return True
    except:
        pass 
    return False
```

```python
NaturezaDespesaDetalhadaLimpa = []
```

```python
for lista in NaturezaDespesaDetalhada:
    if is_number(lista[0]):
        listaLimpa = []
        listaLimpa.append(lista[0])
        listaLimpa.append(lista[4])
        #for elemento in lista:
            #if elemento != '' and elemento != 'D'and elemento != '"':
                #listaLimpa.append(elemento)                               
        NaturezaDespesaDetalhadaLimpa.append(listaLimpa)
print( NaturezaDespesaDetalhadaLimpa)
```

```python
NaturezaDespesaDetalhadaCompleta = []
```

```python
for linha in NaturezaDespesaDetalhadaLimpa:    
    linha.append(linha[0][0])
    linha.append(linha[0][1])
    linha.append(linha[0][2:4])
    linha.append(linha[0][6:])
    NaturezaDespesaDetalhadaCompleta.append(linha)
print(NaturezaDespesaDetalhadaCompleta)
```

```python
for registro in NaturezaDespesaDetalhadaCompleta:
    engine.execute('INSERT INTO tbl_GDFNaturezaDespesaDetalhada (ndf_Codigo, ndf_Descricao, ndf_CategoriaEconomica, ndf_GrupoDespesa, ndf_ModalidadeAplicacao, ndf_ElementoDespesa) values ("'+registro[0]+'","'+registro[1]+'","'+registro[2]+'","'+registro[3]+'","'+registro[4]+'","'+registro[5]+'");')
```
