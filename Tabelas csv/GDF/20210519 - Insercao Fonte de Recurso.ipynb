```python
import pandas as pd
import csv
import mysql.connector
```

```python
fonte = open('Fonte.CSV')
```

```python
fonteReadeble = csv.reader(fonte)
```

```python
fonteLista = list(fonteReadeble)
```

```python
def is_number(num):
    try:
        float(num)
        return True
    except:
        pass 
    return False
```

```python
fonteLimpa = []
```

```python
for lista in fonteLista:
    if is_number(lista[0]):
        listaLimpa = []
        listaLimpa.append(lista[0])
        listaLimpa.append(lista[3])
        listaLimpa.append(lista[6])
        listaLimpa.append(lista[10])
        listaLimpa.append(lista[11])
        listaLimpa.append(lista[0][0:3])
        #for elemento in lista:
            #if elemento != '' and elemento != 'D'and elemento != '"':
                #listaLimpa.append(elemento)                               
        fonteLimpa.append(listaLimpa)
print(fonteLimpa)
```

```python
with open("fonteRecurso.csv", "w") as f:
    writer = csv.writer(f)
    writer.writerows(fonteLimpa)
```
