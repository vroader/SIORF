```python
import pandas as pd
import sqlalchemy
import csv
```

```python
engine = sqlalchemy.create_engine('mysql+pymysql://siorf:PMseis06@mysql.siorf.kinghost.net:3306/siorf')
```

```python
Alinhamento = open('tbl_AlinhamentoEstratégico2.csv', encoding = 'utf-8')
```

```python
Alinhamento_Readeble = csv.reader(Alinhamento, delimiter = ';')
```

```python
listAlinhamento = list(Alinhamento_Readeble)
```

```python
del listAlinhamento[0]
listAlinhamento
```

```python
Estrategia = pd.read_sql_table('tbl_Estrategia',engine)
```

```python
Estrategia 
```

```python
from pandas import DataFrame
dflistAlinhamento = DataFrame(listAlinhamento,columns=['Iniciativa', 'Descricao', 'Estrategia'])
```

```python
dflistAlinhamento
```

```python
dfAlinhamento = dflistAlinhamento.merge(Estrategia, left_on='Estrategia', right_on='est_Codigo', how = 'right')
dfAlinhamento
```

```python
dfAlinhamentoFinal = dfAlinhamento[['Iniciativa', 'Id_Estrategia', 'Descricao']]
```

```python
list_AlinhamentoFinal = dfAlinhamentoFinal.values.tolist()
list_AlinhamentoFinal
```

```python
#Transformando dados 'int' em string'
IniciativaEstrategica = []
for lista in list_AlinhamentoFinal:
    linha = []
    linha.append(lista[0].strip('.'))
    linha.append(str(lista[1]))
    linha.append(lista[2])
    IniciativaEstrategica.append(linha)
print(IniciativaEstrategica)
```

```python
for registro in  IniciativaEstrategica:
    engine.execute('INSERT INTO tbl_IniciativaEstrategica (ini_Codigo, ini_Estrategia, ini_Descricao) values ("'+registro[0]+'","'+registro[1]+'","'+registro[2]+'");')
```
